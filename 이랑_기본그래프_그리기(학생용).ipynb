##🎨그래프에 범례 넣기
* 그래프가 _____하는 바를 구별할 수 있도록 표시해놓는 ______
* 범례는 보통 두 개 이상의 데이터를 표시할 때 사용합니다.

📌label 속성의 레이블 값으로 원하는 문자열을 넣어주고,
그래프를 그리기 전에 legend()함수를 실행시키면 레이블 값이 범례로 나타난다.

*두 가지 범례를 넣으려면 plot함수를 2개 사용한다.


````python
#범례를 표시하는 코드를 작성하세요.

import matplotlib.pyplot as plt
plt.title('plottiong')

plt.plot([10,20,30,40], label= 'asc')  #증가를 뜻하는 asc범례
plt.plot([40,30,20,10], label = 'desc') #감소를 뜻하는 desc 범례
plt.legend()

plt.show()
````

*loc의 값은 0~10까지 넣을 수 있다.

```

```



````python
#범례 위치 아래쪽으로 지정하는 코드를 작성하시오.
import matplotlib.pyplot as plt
plt.title('plottiong')

plt.plot([10,20,30,40], label = 'asc')  #증가를 뜻하는 asc범례
plt.plot([40,30,20,10], label = 'desc') #감소를 뜻하는 desc 범례
plt.legend(loc =  9)

plt.show()
````

### ▶데이터를 리스트에 저장하기

````python
#데이터를 리스트에 저장하기
#최고 기온 데이터를 날짜 순으로 저장하고자 result라는 리스트 생성
#최고 기온데이터가 존재하는 날의 최고 기온 데이터를 result 변수에 저장

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding="euc-kr")
data = csv.reader(f)
next(data)
result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] !='':  #최고 기온 데이터 값이 존재한다면 result리스트에 최고 기온 값 추가
    result.append(float(row[-1]))
print(result)
````

##2)기온데이터를 히스토그램으로 표현하기

````python
#최고 기온데이터를 히스토그램으로 표현하기

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
  if row[-1] !='':
    result.append(float(row[-1]))

plt.hist(result, bins=100, color = 'r')  #result의 결과를 100개의 가로축 구간 개수로 설정하여 히스토그램 그리기
plt.show()
````

````python
#8월 데이터만 히스토그램으로 표현하기
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []

for row in data :
  month =row[0].split('-')[1]                                              #-로 구분된 값 중 2번째 값을 month에 저장
  if row[-1] !='':
    if month =='08':
      aug.append(float(row[-2]))

plt.hist(aug, bins=100, color = 'r')
plt.show()
````

````python
#1월과 8월의 데이터를 히스토그램으로 시각화

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan =[]

for row in data :
  month =row[0].split('-')[1]     #-로 구분된 값 중 2번째 값을 month에 저장
  if row[-2] !='':
    if month =='08':                 #8월이면
      aug.append(float(row[-1]))
    if month =='01':                 #1월이면
      jan.append(float(row[-1]))

plt.hist(aug, bins=100, color = 'r', label = 'aug')        #8월 데이터를 히스토그램으로 출력, 색상은 빨간색, 레이블은 aug로 설정한다.
plt.hist(jan, bins=100, color = 'b', label = 'jan')        #1월 데이터를 히스토그램으로 출력, 색상은 파란색, 레이블은 jan로 설정한다.
plt.legend()
plt.show()
````

##3)기온데이터를 상자 그림으로 표현하기

* 상자그림은 가공하지 않은 데이터를 그대로 이용하는 것이 아니라, 자료에서 얻어낸 ,최댓값, 최솟값, 상위 1/4, 중앙, 3/4에 위치한 값을 보여주는 그래프

````python
import random
import matplotlib.pyplot as plt

result = []
for i in range(13):                          #13번 반복
  result.append(random.randint(1,1000))     #1~1000개의 값을 임의로 추출하여 추가하기
print(sorted(result))

plt.boxplot(result)   #상자그림으로 출력하기
plt.show()
````

````python
#정확한 위치 값이 알고 싶다면? percentile( )함수 사용하기!

import numpy as np
result = np.array(result)
print("1/4 : " +str(np.percentile(result,25)))
print("2/4 : " +str(np.percentile(result,50)))
print("3/4 : " +str(np.percentile(result,75)))
````

````python
#서울의 최고 기온데이터를 상자그림으로 그리기

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
  if row[-1] !='':
    result.append(float(row[-1]))   #서울의 최고기온값을 실수형태로 변환

plt.boxplot(result)                   #결과값을 상자그림으로 표현
plt.show()
````

````python
import numpy as np
result = np.array(result)
print("1/4 : " +str(np.percentile(result,25)))
print("2/4 : " +str(np.percentile(result,50)))
print("3/4 : " +str(np.percentile(result,75)))
````

````python
#1월과 8월의 데이터를 상자 그림으로 시각화

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan =[]

for row in data :
  month =row[0].split('-')[1] #-로 구분된 값 중 2번째 값을 month에 저장
  if row[-1] !='':
    if month =='08':
      aug.append(float(row[-1]))               #8월의 최고기온값을 실수형태로 변환 예)'20.1' -> 20.1
    if month =='01':
      jan.append(float(row[-1]))               #1월의 최고기온값을 실수형태로 변환 예)'6.7' -> 6.7

plt.boxplot(aug)                #8월의 데이터를 상자그림으로 표현하기
plt.boxplot(jan)                #1월의 데이터를 상자그림으로 표현하기

plt.show()
````

````python
#8월 일별 기온데이터를 상자 그림으로 표현하기

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = []                                                                  #일별데이터 리스트 생성
for i in range(31):
  day.append([])                                                         #day리스트 내 31개 리스트 추가하기

for row in data :
    if row[-1] !='':
      if row[0].split('-')[1] =="08" :                                    #8월이라면
        day[int(row[0].split('-')[2])-1].append(float(row[-1]))           #최고 기온 값 저장

plt.style.use('ggplot')                                                   #그래프 스타일 지정
plt.figure(figsize=(10,5), dpi = 300)                                     #그래프 크기 수정
plt.boxplot(day, showfliers=False)                                #아웃라이어 값 생략
plt.show()
````
